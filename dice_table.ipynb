**Dice Roll Simulation with NumPy & Pandas**

This script simulates rolling a six-sided die 1,000 times. It generates random data, processes it into discrete "faces," and summarizes the results in a formatted table to show the distribution of rolls.

```python
import numpy as np
import pandas as pd

n = 1000

# Generating 1000 random numbers between 0 and 1
r = np.random.rand(n)

# Generating 1000 faces, using the floor function to map to dice faces (1–6)
faces = np.floor(r * 6) + 1

# Count frequencies
unique_faces, counts_of_faces = np.unique(faces, return_counts=True)

# Datframe creation (notice the dictionary inside)
dice_table = pd.DataFrame({
    "Face": unique_faces.astype(int),
    "Frequency": counts_of_faces
})

# Adding the  Percentage column in the above dataframe
dice_table["Percentage"] = ((dice_table["Frequency"] / n) * 100).round(1)

dice_table
```
